```python
import pandas as pd
from IPython.display import display, HTML
```

```python
df = pd.read_excel('Customer Behaviors (Responses).xlsx')
df.columns = df.columns.str.replace("คุณมีความสนใจในสิ่งเหล่านี้มากน้อยเพียงใด", "interest")
df.columns = df.columns.str.replace("คุณบริโภคสิ่งเหล่านี้บ่อยขนาดไหน", "consumption")
df.rename(columns={"นอกเหนือจากรายการดังกล่าวแล้ว คุณมีความสนใจในเรื่องใดอีกบ้าง":"other interest", "คุณเกิดวันที่เท่าไหร่":"dob", "เพศ":"gender"}, inplace=True)
```

```python
df.head()
```

```python
df = df.replace(["ไม่สนใจอย่างมากที่สุด", "ไม่สนใจอย่างมาก", "ไม่สนใจ", "เฉยๆ", "สนใจ", "สนใจอย่างมาก", "สนใจอย่างมากที่สุด"], [-3, -2, -1, 0, 1, 2, 3])
df = df.replace(["แทบไม่ได้บริโภคเลย", "หลายเดือนครั้ง", "เดือนละครั้ง", "เดือนละ 2-3 ครั้ง", "อาทิตย์ละครั้ง", "แทบทุกวัน"], [1, 2, 3, 4, 5, 6])
```

```python
df.head()
```

```python
print ("Interest:")
display(df.corr().loc[:'interest [อาหารอีสาน]', 'interest [อาหารสุขภาพ]': 'interest [เหล้าเบียร์]'])
print ("Consumption:")
display(df.corr().loc['consumption [อาหารญี่ปุ่น]':'consumption [อาหารอีสาน]', 'consumption [อาหารสุขภาพ]': 'consumption [เหล้าเบียร์]'])
```
